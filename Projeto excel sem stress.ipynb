```python
#Ler a planilha do mês seguinte para que se complete com a atual 
vendas_jun_df = pd.read_excel("C://Users/bruno/Downloads/Lista de 1 mês depois completo.xlsx")
#display(vendas_jun_df)

vendas_df = pd.concat([vendas_df,vendas_jun_df],axis = 1)
display(vendas_df)
```

```python
#Excluir colunas específicas 
vendas_df = vendas_df.drop("Imposto",axis = 1)
display(vendas_df)
```

```python
#Excluir linhas específica através de seu número 
vendas_df = vendas_df.drop(112,axis = 0) 
display(vendas_df)
```

```python
vendas_df = vendas_df.fillna(45)
display(vendas_df)
#Preencher os valores vazios com algum número 
```

```python
# Remover colunas que contenham pelo menos um valor nulo
vendas_df = vendas_df.dropna(axis=1)
display(vendas_df)
```

```python
#Deletar linhas que possuem pelo menos um valor vazio 
vendas_df = vendas_df.dropna()
display(vendas_df)
```

```python
#Preencher valor vazio com o último valor de cima dele 
vendas_df = vendas_df.ffill()
display(vendas_df)
```

```python
transacoes_produto = vendas_df['Produto'].value_counts()
display(transacoes_produto)
#Quantidade de transações de cada produto
```

```python
#Somar os valores finais dos produtos para ver seu lucro final 
faturamento_produto = vendas_df[['Produto', 'Valor Final']].groupby('Produto').sum() 
display(faturamento_produto)
```
